```python
import numpy as np
```

###입력텐서와 타깃텐서로 이루어진 훈련 데이터를 정의

```python
input_tensor=np.random.rand(100,50) #100개의 데이터와 50개의 피쳐 ex.100명의 학생들&50개의 특성
target_tensor=np.random.randint(low=0,high=10,size=100) #0~9사이까지의 100건의 데이터를 끄낸다
```

##### 입력과 타킷을 매핑하는층으로 이루어진 네트워크(모델)을 정의

```python
from tensorflow.keras import models #인공신경망을 통해 만든 하나의 전체 서비스 (input넣으면 output으로 나옴)
from tensorflow.keras import layers #모델안에 채워넣게된 레이어들 (몇개들의 레이어들 넣고,각각의 레이어들에게 몇 개의 퍼셉트론들을 넣을것인지)
```

```python
model=models.Sequential() #Sequential의 가장 기본적인 모델은 클래스라고 생각
model.add(layers.Dense(units=32,activation="relu",input_shape=(50,))) 
model.add(layers.Dense(units=10,activation="softmax")) #unit=10인 이유는 0~9까지 10개의 클래스를 가지기 때문
#이 문제가 classfication 문제라서 softmax로 정의
```

######손실함수,옵티마이저(gradient descent할지 어떤 알고리즘을 할지),모니터링하기위한 성능지표(Accuracy,Precision,Recall) 선택하여 학습과정을 설정

```python
from tensorflow.keras import optimizers
```

```python
model.compile(optimizer=optimizers.RMSprop(lr=0.001),loss="mse",metrics=["accuracy"])
#optimizer=optimizers.RMSprop에서 우리가 아는 GD가 아닌 다른 알고리즘을 사용
#그 알고리즘의 학습률을 lr=0.001로 선택
#loss function은 mse로 선택
#성능지표는 accuracy로 선택
```

####훈련데이터에 대해 모델객체의 fit()_method를 실행

```python
model.fit(x=input_tensor,y=target_tensor,batch_size=10,epochs=10)
#x=input_tensor는 100개의 데이터고 50개의 속성정보를 가지고 있음
#y=target_tensorsms 100건의 데이터인데 0~9사이의 값인 것
#epoch 몇 번 학습시킬지, batch_size는 한번에 몇번을 넣어서 학습시킬지
```
